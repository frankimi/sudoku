## Help Functions 

```python
def block(block_i,block_j):
    return table.iloc[3*block_i:3*(block_i+1),3*block_j:3*(block_j+1)]

def row(i):
    return table.iloc[i,:]

def column(j):
    return table.iloc[:,j]
```

```python
def possible_values(i,j):
    #print(table.iloc[i,j])
    block_i = int(i/3)
    block_j = int(j/3)
    #display(block(block_i,block_j))
    block_values = list(set(np.reshape(block(block_i,block_j).values,9)))
    #print(block_values)
    
    
    
    row_values = list(set(row(i).values))
    #print(row_values)
    
    column_values = list(set(column(j).values))
    #print(column_values)
    
    possible_values_list = list(range(1,10))
    possible_values_list = [x for x in possible_values_list if x not in row_values]
    possible_values_list = [x for x in possible_values_list if x not in column_values]
    possible_values_list = [x for x in possible_values_list if x not in block_values]
    
    #print(possible_values_list)
    return possible_values_list
    
```

```python
def block_unique_possible_value(block_i,block_j):
    unique_list = []

    for i in range(block_i*3, block_i*3 + 3):
        for j in range(block_j*3, block_j*3 + 3):
            if table.iloc[i,j] == 0:
                if len(unique_list) == 0:
                    unique_list = possible_values(i,j)
                else:
                    unique_list = unique_list + possible_values(i,j)

    unique_list = list(set([i for i in unique_list if unique_list.count(i)<=1]))
    return unique_list
```

```python
def row_unique_possible_value(i):
    unique_list = []
    for j in range(0,9):
        if table.iloc[i,j] == 0:
            if len(unique_list) == 0:
                unique_list = possible_values(i,j)
            else:
                unique_list = unique_list + possible_values(i,j)

    unique_list = list(set([i for i in unique_list if unique_list.count(i)<=1]))
    return unique_list
```

```python
def column_unique_possible_value(j):
    unique_list = []
    for i in range(0,9):
        if table.iloc[i,j] == 0:
            if len(unique_list) == 0:
                unique_list = possible_values(i,j)
            else:
                unique_list = unique_list + possible_values(i,j)

    unique_list = list(set([i for i in unique_list if unique_list.count(i)<=1]))
    return unique_list
```

```python
def solver():
    
    # Method 1
    for i in range(0,9):
        for j in range(0,9):
            if table.iloc[i,j] == 0:
                #print(table.iloc[i,j],'At',(i,j), 'possible values:', possible_values(i,j))
                if len(possible_values(i,j)) == 1:
                    table.iloc[i,j] = possible_values(i,j)[0]
    
    # Method block
    for block_i in range(0,3):
        for block_j in range(0,3):
            values = block_unique_possible_value(block_i,block_j)
            
            for i in range(block_i*3, block_i*3 + 3):
                for j in range(block_j*3, block_j*3 + 3):
                    if len(values) == 0:
                        break
                    p_v = possible_values(i,j)
                    if table.iloc[i,j] == 0:
                        for x in values:
                            if x in p_v:
                                table.iloc[i,j] = x
                                values.remove(x)
                                break
    # Method row
    for i in range(0,9):
        values = row_unique_possible_value(i)
        for j in range(0,9):
            p_v = possible_values(i,j)
            if table.iloc[i,j] == 0:
                for x in values:
                    if x in p_v:
                        table.iloc[i,j] = x
                        values.remove(x)
                        break
                    
    # Method column                
    for j in range(0,9):
        values = column_unique_possible_value(j)
        for i in range(0,9):
            p_v = possible_values(i,j)
            if table.iloc[i,j] == 0:
                for x in values:
                    if x in p_v:
                        table.iloc[i,j] = x
                        values.remove(x)
                        break
            
               
                
```

```python
def play_Sudoku():
    
    # Solve
    last_zeros = 81
    while 0 in list(set(np.reshape(table.values,81))):
        zeros = len([x for x in list(np.reshape(table.values,81)) if x < 1])
        solver()
        if zeros < last_zeros:
            last_zeros = zeros
        else:
            print('Cannot solve')
            break
    print('Result:')
    display(table)
    
```
